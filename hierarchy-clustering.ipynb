```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../jupyter/environment/cluster_mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_dummies=pd.get_dummies(df.drop('name',axis=1))
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_X=scaler.fit_transform(df_dummies)
```

```python
scaled_X
```

```python
scaled_df=pd.DataFrame(scaled_X,columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8),dpi=300)
sns.clustermap(scaled_df,cmap='viridis', col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=4,affinity='cosine')
```

```python
cluster_labels=cluster.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',y='horsepower',hue=cluster_labels,palette='viridis')
```

```python
np.sqrt(len(scaled_df.columns))
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=2)
```

```python
labels=model.fit_predict(scaled_df)
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
from scipy.cluster import hierarchy
```

```python
model.children_
```

```python
linkage_matrix=hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10),dpi=200)
dendro=dendrogram(linkage_matrix,truncate_mode='lastp',p=11)
plt.xticks(rotation=90);
```
