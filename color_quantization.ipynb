```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(img_2d)
```

```python
labels
```

```python
rgb=np.round(model.cluster_centers_,0).astype(int)
```

```python
rgb[labels]
```

```python
rgb[3]
```

```python
quantized_img=np.reshape(rgb[labels],(h,w,c))
```
