```python
pixles=df.drop('number_label',axis=1)
```

```python
sns.heatmap(number,annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_pixles=scaler.fit_transform(pixles)
```

```python
scaled_pixles=pd.DataFrame(scaled_pixles,columns=pixles.columns)
```

```python
scaled_pixles
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model=PCA(n_components=2)
```

```python
pca_2=pca_model.fit_transform(scaled_pixles)
```

```python
pca_2
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x=pca_2[:,0],y=pca_2[:,1],hue=df.number_label,palette='Set1')
```

```python
pca_3=PCA(n_components=3)
```

```python
pca_pixles=pca_3.fit_transform(scaled_pixles)
```

```python
from mpl_toolkits import mplot3d
```

```python
ax=plt.axes(projection='3d')
```

```python
plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.scatter3D(xs=pca_pixles[:,0],ys=pca_pixles[:,1],zs=pca_pixles[:,2]
             ,c=df.number_label,)
plt.show()
```
