```python
cancer_dict=load_breast_cancer()
```

```python
scaled_X_df=pd.DataFrame(scaled_X,columns=df.columns)
```

```python
scaled_X_df
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)
```

```python
model.fit(scaled_X_df)
```

```python
pc=model.transform(scaled_X_df)
```

```python
sns.scatterplot(x=pc[:,0],y=pc[:,1],hue=cancer_dict['target'])
```

```python
df_com=pd.DataFrame(model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
plt.figure(figsize=(20,8),dpi=200)
sns.heatmap(df_com,annot=True)
```

```python
np.sum(model.explained_variance_ratio_)
```

```python
pc_30=PCA(n_components=30)
```

```python
 pc_30.fit_transform(scaled_X_df)
```

```python
np.sum(pc_30.explained_variance_ratio_)
```

```python
explained_var=[]
for n in range(1,30):
    pca=PCA(n_components=n)
    pca.fit(scaled_X_df)
    explained_var.append(np.sum(pca.explained_variance_ratio_))
```

```python
explained_var
```

```python
plt.plot(range(1,30),explained_var)
plt.xlabel('n_components')
plt.ylabel('variance explained')
```
