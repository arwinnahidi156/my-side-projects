```python
def make_full_matrix(sparse_matrix,nr_words,doc_idx=0,word_idx=1,label_idx=2,freq_idx=3):
    col_names=['doc_id'] + ['label'] + list(range(nr_words))
    index_names=np.unique(sparse_matrix[:,0])
    full_matrix=pd.DataFrame(index=index_names,columns=col_names)
    full_matrix.fillna(0,inplace=True)
    for i in range(sparse_matrix.shape[0]):
        doc_num=sparse_matrix[i][doc_idx]
        word_id=sparse_matrix[i][word_idx]
        label=sparse_matrix[i][label_idx]
        occurence=sparse_matrix[i][freq_idx]
        full_matrix.at[doc_num,'doc_id']=doc_num
        full_matrix.at[doc_num,'label']=label
        full_matrix.at[doc_num,word_id]=occurence
        
    return full_matrix
        
        
    
    
    
    
    
    
    
    
    
    
    
    
    
```
